```python
import pandas as pd
import os
import pymongo
```

```python
# Import CSV file# 1
nba_file = os.path.join('data','nba_2017_players_stats_combined.csv')
nba_df = pd.read_csv(nba_file, encoding='iso-8859-1')
```

```python
nba_df.head()
```

```python
# Drop duplicate column MP
nba_df.drop('MP', inplace=True, axis=1)
```

```python
nba_df.head()
```

```python
# Rename the columns of nba_df: nba_df_tidy
nba_df_tidy = nba_df.rename(columns = {'W': 'WINS'}, inplace = False)
nba_df_tidy.head()
```

```python
nba_df_tidy.info()
```

```python
# Drop duplicate column index
nba_df_tidy.drop('Unnamed: 0', inplace=True, axis=1)
nba_df_tidy.head()
```

```python
# Import CSV file# 2
nba_file_2 = os.path.join('data','nba_2017_twitter_players.csv')
nba_twitter_df = pd.read_csv(nba_file_2, encoding='iso-8859-1')
nba_twitter_df.head()
```

```python
nba_twitter_df.info()
```

```python
# Drop Players with no Twitter data
nba_twitter_df.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=True
)
```

```python
nba_twitter_df.info()
```

```python
# Import CSV file# 3
nba_file_3 = os.path.join('data','nba_2017_att_val_elo_with_cluster.csv')
nba_attendance_df = pd.read_csv(nba_file_3, encoding='iso-8859-1')
nba_attendance_df.head()
```

```python
# Drop duplicate column index
nba_attendance_df.drop('Unnamed: 0', inplace=True, axis=1)
nba_attendance_df.head()
```

```python
# Rename the columns of nba_attendance_df: nba_attendance_df_tidy
nba_attendance_df_tidy = nba_attendance_df.rename(columns = {'GMS': 'GAMES', 'TOTAL':'TOTAL_ATTENDANCE','CONF': 'CONFERENCE', 'cluster': 'CLUSTER'}, inplace = False)
nba_attendance_df_tidy.head()
```

```python
# Import CSV file# 4
nba_file_4 = os.path.join('data','nba_2017_endorsements.csv')
nba_endorsements_df = pd.read_csv(nba_file_4, encoding='iso-8859-1')
nba_endorsements_df.head()
```

```python
# Rename the columns of nba_endorsements_df: nba_endorsements_df_tidy
nba_endorsements_df = nba_endorsements_df.rename(columns = {'ï»¿NAME': 'PLAYER', 'SALARY': 'SALARY_MILLIONS', 'ENDORSEMENT': 'ENDORSEMENT_MILLIONS'}, inplace = False)
nba_endorsements_df.head()
```

```python
nba_endorsements_df.info()
```

```python
#nba_endorsements_df['SALARY_MILLIONS'] = pd.to_numeric(nba_endorsements_df.SALARY_MILLIONS,errors='coerce')
nba_endorsements_df['SALARY_MILLIONS'] = nba_endorsements_df['SALARY_MILLIONS'].str.replace('$', '')
nba_endorsements_df['SALARY_MILLIONS'] = nba_endorsements_df['SALARY_MILLIONS'].str.replace(',', '')
nba_endorsements_df['SALARY_MILLIONS'] = nba_endorsements_df['SALARY_MILLIONS'].astype(float)
nba_endorsements_df.head()

```

```python
nba_endorsements_df['ENDORSEMENT_MILLIONS'] = nba_endorsements_df['ENDORSEMENT_MILLIONS'].str.replace('$', '')
nba_endorsements_df['ENDORSEMENT_MILLIONS'] = nba_endorsements_df['ENDORSEMENT_MILLIONS'].str.replace(',', '')
nba_endorsements_df['ENDORSEMENT_MILLIONS'] = nba_endorsements_df['ENDORSEMENT_MILLIONS'].astype(float)
nba_endorsements_df.head()
```

```python
nba_endorsements_df['SALARY_MILLIONS'] = nba_endorsements_df['SALARY_MILLIONS'] / 1000000
nba_endorsements_df.head()
```

```python
nba_endorsements_df['ENDORSEMENT_MILLIONS'] = nba_endorsements_df['ENDORSEMENT_MILLIONS'] / 1000000
nba_endorsements_df.head()
```

```python
nba_endorsements_df['SALARY_MILLIONS'] = nba_endorsements_df['SALARY_MILLIONS'].round(2)
nba_endorsements_df.head()
```

```python
nba_endorsements_df.dtypes
```

```python
# Import CSV file# 5
nba_file_5 = os.path.join('data','nba_2017_nba_players_with_salary.csv')
nba_salary_df = pd.read_csv(nba_file_5, encoding='iso-8859-1')
nba_salary_df.head()
```

```python
nba_salary_df.info()
```

```python
# Create new dataframe
salary_df=nba_salary_df[['PLAYER', 'SALARY_MILLIONS']]
salary_df.head()
```

```python
salary_df.info()
```

```python
salary_df.drop_duplicates(subset ="PLAYER", 
                     keep = False, inplace = True) 
```

```python
salary_df.info()
```

```python
#Merge SALARY_MILLIONS for player to nba_df_tidy
nba_df_merge_col = pd.merge(nba_df_tidy, salary_df, on='PLAYER')
nba_df_merge_col.head()
```

```python
nba_df_merge_col.info()
```

```python
nba_df_merge_col.head()
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.nba_db
collection1 = db.stats
collection2 = db.twitter
collection3 = db.attendance
```

```python
nba_df_merge_col.reset_index(inplace=True)
stats_data = nba_df_merge_col.to_dict("records")
collection1.insert_many(stats_data)
```

```python
nba_twitter_df.reset_index(inplace=True)
twitter_data = nba_twitter_df.to_dict("records")
collection2.insert_many(twitter_data)
```

```python
nba_attendance_df_tidy.reset_index(inplace=True)
attendance_data = nba_attendance_df_tidy.to_dict("records")
collection3.insert_many(attendance_data)
```
