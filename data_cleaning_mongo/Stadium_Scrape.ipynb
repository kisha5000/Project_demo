```python
#import dependencies
import pandas as pd
import pymongo
import requests
from bs4 import BeautifulSoup
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.nba_db
collection = db.Stadium_data
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_National_Basketball_Association_arenas"
wiki_url= requests.get(URL)
soup = BeautifulSoup(wiki_url.text, 'html.parser')
tables = soup.find_all('table', class_='sortable')
arena_data=pd.read_html(str(tables))
arena_data
```

```python
nba_arenas=arena_data[0]
nba_arenas.head()
```

```python
nba_arenas_df=nba_arenas.drop(['Image','Opened','Ref(s)'],axis=1)
nba_arenas_df
```

```python
from opencage.geocoder import OpenCageGeocode
key = "06cbb655277e4d4a9b075579925aa006"  # get api key from:  https://opencagedata.com
geocoder = OpenCageGeocode(key)
```

```python
list_lat = []   # create empty lists

list_long = []


for index, row in nba_arenas_df.iterrows(): # iterate over rows in dataframe

#     City = row['City']
#     State = row['State']       
    query = row['Location']  

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

	
# create new columns from lists    

nba_arenas_df['lat'] = list_lat   

nba_arenas_df['lon'] = list_long
nba_arenas_df
```

```python
Stadium_data=nba_arenas_df.rename(columns = {'Team(s)':'TEAM' }, inplace = False)
Stadium_data.head()
```

```python
Stadium_data.reset_index(inplace=True)
arena_dict = Stadium_data.to_dict("records")
# Insert collection
collection.insert_many(arena_dict)
```
